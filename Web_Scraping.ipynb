```python
import requests
import pandas as pd
from bs4 import BeautifulSoup


data = []

p = 1
while p <= 14:
    url = f'https://www.car.gr/used-cars/peugeot/307.html?category=15001&make=13198&model=14785&pg={p}'
    p += 1
    print(f"Fetching: {url}")

    r = requests.get(url)
    page_text = r.text
    soup = BeautifulSoup(page_text, 'html.parser')

    price_elements = soup.find_all('div', class_="price-tag")
    title_elements = soup.find_all('h2', class_='title')
    klm_elements = soup.find_all('div', class_='tw-text-grey-600')

    # Convert data into lists
    df_price = [i.text.strip() for i in price_elements]
    df_title = [j.text.strip() for j in title_elements]
    df_klm = [x.text.strip() for x in klm_elements]

    # Add data as dictionaries into the list 'data'
    for i in range(len(df_price)):
        data.append({
            'Title': df_title[i] if i < len(df_title) else None,
            'Price': df_price[i] if i < len(df_price) else None,
            'Kilometers': df_klm[i] if i < len(df_klm) else None
        })

# Create a Pandas DataFrame from the list 'data'
df = pd.DataFrame(data)

# Save the data like a csv file
df.to_csv('car_listings.csv', index=False, encoding='utf-8')

print("The data are saved in 'car_listings.csv'.")
```
